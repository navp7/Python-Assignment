### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular data visualization library in Python. It provides a wide range of tools for creating various types of plots, charts, and graphs. Matplotlib is widely used in the field of data analysis, scientific research, and data visualization tasks.

Here are a few reasons why Matplotlib is used:

1. **Creating Publication-Quality Visualizations:** Matplotlib allows users to create high-quality visualizations suitable for publication or presentation purposes. It provides extensive customization options to control every aspect of the plot, such as colors, labels, fonts, line styles, and more.

2. **Wide Range of Plot Types:** Matplotlib supports a diverse set of plot types, enabling users to represent data in different ways. It can handle line plots, scatter plots, bar plots, histograms, pie charts, box plots, and many more.

3. **Integration with NumPy and Pandas:** Matplotlib seamlessly integrates with other scientific computing libraries like NumPy and Pandas, allowing users to visualize data stored in arrays or DataFrames easily.

4. **Interactive Plotting and Embeddability:** Matplotlib can be used to create interactive plots that respond to user interactions such as zooming, panning, and hovering. It is also possible to embed Matplotlib plots in graphical user interfaces (GUIs) or web applications.

5. **Extensibility and Community Support:** Matplotlib is highly extensible and can be customized and extended through various libraries and toolkits. It has a large and active community, which means there are plenty of resources, examples, and user-contributed packages available to enhance its functionality.

Five plots that can be created using the `pyplot` module of Matplotlib are:

1. **Line Plot:** A line plot shows the relationship between two variables by connecting data points with straight lines.
2. **Scatter Plot:** A scatter plot displays individual data points as markers on a coordinate grid to show the correlation between two variables.
3. **Bar Plot:** A bar plot represents categorical data using rectangular bars of varying heights, where the length of each bar corresponds to the value being plotted.
4. **Histogram:** A histogram is used to visualize the distribution of a continuous variable by dividing the data into bins and displaying the frequency or count of values in each bin.
5. **Pie Chart:** A pie chart represents categorical data as slices of a circle, where the size of each slice corresponds to the proportion or percentage of the data category it represents.

These are just a few examples, and Matplotlib offers many more types of plots to suit different visualization needs.

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

### Note: Also add title, xlabel, and ylabel to the plot.

#### Ans: A scatter plot displays individual data points as markers on a coordinate grid to show the correlation between two variables.

```python
# Code
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("This is X-Axis")
plt.ylabel("This is Y-Axis")
plt.title("Random Scatter Plot")
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

### Use the following data:

##### import numpy as np
##### For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
##### For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
##### For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
##### For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

#### The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to arrange multiple plots or axes in a grid-like structure, making it easier to compare and visualize different aspects of the data.

#### The subplot() function takes three arguments: nrows, ncols, and index. These arguments specify the number of rows and columns in the grid and the index of the specific subplot you want to create.

```python
# Code

import matplotlib.pyplot as plt
import numpy as np

plt.subplot(2, 2, 1)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x,y)
plt.title("1st Data")

plt.subplot(2, 2, 2)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x,y)
plt.title("2nd Data")

plt.subplot(2, 2, 3)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x,y)
plt.title("3rd Data")

plt.subplot(2, 2, 4)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x,y)
plt.title("4th Data")

plt.tight_layout()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

#### import numpy as np
#### company = np.array(["Apple", "Microsoft", "Google", "AMD"])
#### profit = np.array([3000, 8000, 1000, 10000])

#### Ans. A bar plot represents categorical data using rectangular bars of varying heights, where the length of each bar corresponds to the value being plotted.

```python
#code

import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.subplot(2,1,1)
plt.bar(company , profit)
plt.xlabel('Company')
plt.ylabel('Profit')

plt.subplot(2,1,2)
plt.barh(company , profit)
plt.ylabel('Company')
plt.xlabel('Profit')

plt.tight_layout()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

### box1 = np.random.normal(100, 10, 200)
### box2 = np.random.normal(90, 20, 200)

#### Ans: Matplotlib's boxplot mainly provides a graphical summary of a data set with features such as minimum, first quartile, median, third quartile, and maximum.

#### The "Whisker Plot" is another name for the "Box Plot."
#### The whiskers are the two lines that run from the minimum to the lower quartile (the box's beginning) and then from the higher quartile (the box's end) to the maximum.
#### In the box plot, the box is drawn from the first to third quartiles, with a vertical line running through it at the median.
#### The x-axis of a box plot represents the data to be plotted, while the y-axis represents the frequency distribution.
#### Box plots are useful for visualizing the distribution of numerical values in a field. They come in handy for making comparisons across categorical variables and spotting outliers, if any exist in a dataset.

```python
# Code

import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
box_m = [box1, box2]
figure = plt.figure(figsize =(8,6))  
plt.boxplot(box_m)
```
