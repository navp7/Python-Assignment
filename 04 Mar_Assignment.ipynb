### Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import plotly.graph_objects as go
import seaborn as sns

df = sns.load_dataset('titanic')
fig = go.Figure()
fig.add_trace(go.Scatter(x = df['age'] , y =df['fare'] , mode = 'markers', name ='Scatter Plot of Age and Fare'))
fig.show()
```

### Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
import plotly.graph_objects as go
import plotly.express as px

df = px.data.tips()
fig = go.Figure()
fig.add_trace(go.Histogram(x=df['sex'],  y=df['total_bill']))

# Add shape pattern w.r.t smoker column:
pattern_shapes = {'Yes': 'x', 'No': 'circle'}
for smoker, shape in pattern_shapes.items():
    df_smoker = df[df['smoker'] == smoker]
    fig.add_trace(go.Scatter(
        x=df_smoker['sex'],
        y=df_smoker['total_bill'],
        mode='markers',
        name=f'Smoker={smoker}',
        marker=dict(symbol=shape, size=10),
        showlegend=True
    ))
# Add colour w.r.t day column:
color_map = {'Thur': 'red', 'Fri': 'black', 'Sat': 'green', 'Sun': 'orange'}
for day, color in color_map.items():
    filtered_data = df[df['day'] == day]
    fig.add_trace(go.Scatter(
        x=filtered_data['sex'],
        y=filtered_data['total_bill'],
        mode='markers',
        name=day,
        marker=dict(color=color, size=10),
    ))



# Update layout
fig.update_layout(
    xaxis=dict(title='Sex'),
    yaxis=dict(title='Total Bill'),
    title='Histogram of Total Bill by Sex',
    showlegend=True,
    legend_title='Smoker', width=1000, height=800)

fig.show()
```

## Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.
## Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the dimensions parameter.

```python
import plotly.express as px

iris = px.data.iris()

fig = px.scatter_matrix(
    iris,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species",
    title="Scatter Matrix Plot of Iris Dataset",

)

fig.update_layout(
    height=600,
    width=800
)

fig.show()
```

### Q5. What is Distplot? Using Plotly express, plot a distplot.

A distplot (distribution plot) is a graphical representation of the distribution of a univariate dataset. It combines a histogram with a kernel density plot to provide a smooth estimate of the probability density function (PDF) of the data. Distplots are useful for visualizing the underlying distribution and can help identify patterns such as peaks, gaps, or skewness in the data.

Using Plotly Express, you can create a distplot using the px.histogram() function, which automatically incorporates the kernel density estimation (KDE) to create the smooth curve.

```python
import plotly.express as px

tips = px.data.tips()

fig = px.histogram(tips, x="total_bill", nbins=30, marginal="rug", title="Distplot of Total Bill")

fig.show()
```
