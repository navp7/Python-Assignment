### Q1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans:

Seaborn is a powerful data visualization library in Python that provides a high-level interface for creating informative and visually appealing statistical graphics. Here are five common plots that can be created using Seaborn, along with their typical use cases:

1. Scatter Plot:
   - Uses: Scatter plots are used to visualize the relationship between two continuous variables. They help in identifying patterns, clusters, or trends in the data. They are also useful for detecting outliers.

2. Line Plot:
   - Uses: Line plots are commonly used to display trends over time or any ordered data. They are effective for visualizing data with a clear progression and can be used to show changes in variables over a specific interval.

3. Bar Plot:
   - Uses: Bar plots are used to compare and display categorical data. They are suitable for showing the distribution or frequency of categories and are helpful in identifying the largest or smallest categories.

4. Histogram:
   - Uses: Histograms are used to visualize the distribution of a single variable. They provide insights into the shape, central tendency, and spread of the data. Histograms are particularly useful for identifying skewness and outliers.

5. Heatmap:
   - Uses: Heatmaps are used to display a matrix of values as a color-encoded grid. They are helpful for visualizing the relationships between two variables and are commonly used to show correlation matrices or confusion matrices.


### Q2. Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
#Code

import seaborn as sns
fmri = sns.load_dataset('fmri')
sns.lineplot(data = fmri, x = 'timepoint' , y = 'signal' , hue ='event', style = 'region' )
```

### Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
#Code

import seaborn as sns
import matplotlib.pyplot as plt

d1 = sns.load_dataset('titanic')

plt.figure(figsize=(8, 5))
sns.boxplot(data=d1 , x='pclass', y='age')

plt.figure(figsize=(8, 5))
sns.boxplot(data=d1 , x='pclass', y='fare')
```

### Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
#Code

import seaborn as sns
import matplotlib.pyplot as plt
dmond = sns.load_dataset('diamonds')
plt.figure(figsize = (10,5))
sns.histplot(data = dmond , x = 'price', hue = 'cut', multiple = 'stack' )
```

### Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
#Code

import seaborn as sns

d2 = sns.load_dataset('iris')
sns.pairplot(d2, hue ='species')
```
